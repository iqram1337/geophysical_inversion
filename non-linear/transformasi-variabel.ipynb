```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parameterisasi
x = np.array([1, 2, 3, 4])
y = np.array([1, 2, 3, 5])

# matriks kernel
G1 = np.column_stack([x**0, x])

# inversi 1
m_calc = np.linalg.inv((G1.T).dot(G1)).dot(G1.T).dot(y)
m_calc
```

```python
# parameterisasi
xt = y
yt = x

# matriks kernel
G2 = np.column_stack([xt**0, xt])

# inversi 2
m_calc2 = np.linalg.inv((G2.T).dot(G2)).dot(G2.T).dot(yt)
m_calc2
```

```python
# plotting
fig,(ax1, ax2) = plt.subplots(1, 2, figsize=(12,5)) # menyiapkan dua plot axis

# ax1
ax1.plot(x,y,'ko')
ax1.plot(x, G1.dot(m_calc), "k-")

# ax2
ax2.plot(xt,yt,'ko')
ax2.plot(xt, G2.dot(m_calc2), "k-")
plt.show()
```
